## Feature Encoding and Discretization

1. Apply label encoding to OFFENSE_CODE_GROUP and DAY_OF_WEEK
2. Apply One-Hot Encoding to UCR_PART
3. Apply Discretization to HOUR (0-8 Early Morning , 8-12 Morning 12-16 Afternoon, 16-19 Evening, 19-23 Night) or you can create your own levels.

```python
from numpy import array
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df2["OFFENSE_CODE_GROUP"] = label_encoder.fit_transform(df2["OFFENSE_CODE_GROUP"])
```

```python
df2['DAY_OF_WEEK'] = label_encoder.fit_transform(df2['DAY_OF_WEEK'])
```

```python
df2.head()
```

```python
df2['UCR_PART'].dtypes
```

```python
df2['UCR_PART'] = pd.Categorical(df2['UCR_PART'])
```

```python
dummy = pd.get_dummies(df2['UCR_PART'], prefix = 'UCR_PART')
```

```python
dummy.head()
```

```python
df2 = pd.concat([df2, dummy], axis=1)
```

```python
df2.head()
```

```python
df2['HOUR']  = np.where((df2['HOUR'] <8) ,'Early Morning',
                               np.where((df2['HOUR'] >=8) & (df2['HOUR'] <=12),'Morning',
                               np.where((df2['HOUR']>12) & (df2['HOUR'] <=16),'Afternoon',
                               np.where((df2['HOUR']>16) & (df2['HOUR'] <=19),'Evening',
                                       'Night'))))
```
